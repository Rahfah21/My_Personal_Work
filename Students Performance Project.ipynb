```python
import seaborn as sns
```

```python
# Set the sytle
sns.set(style="whitegrid")

# Initialze the matplotlib figure
fig, axes= plt.subplots(3,2,figsize=(14,18))
# Gender Vs Scores
sns.barplot(x="Gender", y="Math score", data=Student_Performance, ax=axes[0,0])
axes[0,0].set_title("Math Score by Gender")

sns.barplot(x="Gender", y="Reading score", data=Student_Performance, ax=axes[0,1])
axes[0,1].set_title("Reading Score by Gender")

sns.barplot(x="Gender", y="Writing score", data=Student_Performance, ax=axes[1,0])
axes[1,0].set_title("Writing Score by Gender")

# Parential Level of education vs Scores
sns.barplot(x="Parental level of education", y="Math score", data=Student_Performance, ax=axes[1,1])
axes[1,1].set_title("Math Score by Parental Level of Education")
axes[1,1].tick_params(axis='x', rotation=90)

sns.barplot(x="Parental level of education", y="Reading score", data=Student_Performance, ax=axes[2,0])
axes[2,0].set_title("Reading Score by Parental Level of Education")
axes[2,0].tick_params(axis='x', rotation=90)

sns.barplot(x="Parental level of education", y="Writing score", data=Student_Performance, ax=axes[2,1])
axes[2,1].set_title("Writing by Parental Level of Education")
axes[2,1].tick_params(axis='x', rotation=90)

# Adjust the Layout
plt.tight_layout()

# Show the plot
plt.show()
```

# Report 

## Gender and Performance 
The first chart shows that the male students have higher math score compared to female students.
<br>
Second chart shows that the female students perform well in reading than the male students.
<br>
Similarly to reading scores, female students show higher average writing scores compared to the male.

## Parental Level of Education and Performaance 
Students whose perents have a bachelor's degree tend to have the highest math score.
<br> While students with parents who have a master's degree or a bachelor's degree show higher score in both reading and writing.

```python
# Set the sytle
sns.set(style="darkgrid")

# Initialze the matplotlib figure
fig, axes= plt.subplots(3,2,figsize=(14,18))

# Lunch Vs Scores
sns.barplot(x="Lunch", y="Math score", data=Student_Performance, ax=axes[0,0])
axes[0,0,].set_title("Math Score by Lunch")

sns.barplot(x="Lunch", y="Reading score", data=Student_Performance, ax=axes[0,1])
axes[0,1].set_title("Reading Score by Lunch")

sns.barplot(x="Lunch", y="Writing score", data=Student_Performance, ax=axes[1,0])
axes[1,0].set_title("Writing Score by Lunch")

# Test preparation course  vs Scores
sns.barplot(x="Test preparation course", y="Math score", data=Student_Performance, ax=axes[1,1])
axes[1,1].set_title("Math Score by Test preparation course ")
axes[1,1].tick_params(axis='x', rotation=90)

sns.barplot(x="Test preparation course", y="Reading score", data=Student_Performance, ax=axes[2,0])
axes[2,0].set_title("Reading Score by Test preparation course ")
axes[2,0].tick_params(axis='x', rotation=90)

sns.barplot(x="Test preparation course", y="Writing score", data=Student_Performance, ax=axes[2,1])
axes[2,1].set_title("Writing by Test preparation course ")
axes[2,1].tick_params(axis='x', rotation=90)

# Adjust the Layout
plt.tight_layout()

# Show the plot
plt.show()
```

### Lunch and Performance 
Students who recieve standard lunch have higher scores in all the subjects compared to those with free/reduced lunch.

### Test Preparation and Performance 
Students who have completed the test preparation course generally score higher in all courses than those who have not

# Conclusion
1. Male students tend to perform better in math, while female students perform well in both reading and writing courses.
<br>
2. Higher parental education levels correlate with better student performance across all subjects.
<br>
3. Students with standard lunch perform better in math, reading, and writing.
<br>
4. Test preparation courses positively affect the performance in math, reading, and writing.
