# Sales Performance Analysis

### Objectives 
 * To identify trends, pattern, and areas for improvement in sales performance. 

#### Import Some Importance Libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Loading Dataset 

```python
Sales_Performance=pd.read_excel("superstore dataset.xlsx")
Sales_Performance
```

##  Cleaning of Dataset


###### Task to do
* Check the dataset
* Handle missing values
* Remove duplicates
* Format the data

### Checking the data information 

```python
Sales_Performance.info()
```

```python
Sales_Performance.dropna(inplace=True)
Sales_Performance
```

```python
print(' The number of rows after dropping the Non Values is 9789')
```

```python
Sales_Performance=Sales_Performance.drop_duplicates()
Sales_Performance
```

### Changing Data Type
* Order Date
* Ship Date
* Postal Code

```python
Order_Date_Format=pd.to_datetime(Sales_Performance['Order_Date'])
```

```python
Sales_Performance['Order_Date']=Order_Date_Format
```

```python
Ship_Date_Format=pd.to_datetime(Sales_Performance['Ship_Date'])
```

```python
Sales_Performance['Ship_Date']=Ship_Date_Format
```

```python
Sales_Performance['Postal_Code']=Sales_Performance['Postal_Code'].astype(int)
```

```python
Sales_Performance['Sales']=Sales_Performance['Sales'].round(2)
```

```python
Sales_Performance.info()
```

```python
Sales_Performance.head()
```

# Exploratory Data Analysis (EDA) 

```python
Sales_Performance.Sales.describe()
```

### Analyze Sales trends overtime

```python
# Group sales by year.
```

```python
Yearly_Sales=Sales_Performance.groupby(Sales_Performance['Order_Date'].dt.year)['Sales'].sum().reset_index()
Yearly_Sales
```

### Regional Performance Analysis 
   * Evaluate Sales Performance across different regions 

```python
Regional_Sales= Sales_Performance.groupby(Sales_Performance['Region'])['Sales'].sum().reset_index()
Regional_Sales
```

### Sales by Product Category

```python
Product_Category_Sales=Sales_Performance.groupby(Sales_Performance['Category'])['Sales'].sum().reset_index()
Product_Category_Sales
```

### Data Visualization 

#### Line Chart for Sales Trend 

```python
Yearly_Sales.plot(x='Order_Date',y='Sales', figsize=(8,6), marker='o')
plt.title('Year Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

#### Bar Chart for Regional Sales performance 

```python
Regional_Sales.plot(x='Region',y='Sales', kind='bar',figsize=(8,6),width=0.3)
plt.title('Regional Sales Performance')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

#### Pie chart for Product Category

```python
Product_Category_Sales.plot(y='Sales', kind='pie',figsize=(5,5),autopct='%1.1f%%',legend=False,
labels=['Furniture','Office Supplies','Technology'])
plt.title('Product Category Sales Distribution')
plt.show()
```
